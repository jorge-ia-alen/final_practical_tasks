```python
# Defino un diccionario para cada una de las currencies y sus precios por anho
# tambien hago un nuevo data frame para jugarle todo mal
# ultima copy 06-12-2023 formato yakee (mm/dd/yyyy)
aGov = abdo_nueT_copy
cambiosAnhoJPY = {'JPY_2018': 110.28,'JPY_2019': 109.49,'JPY_2020': 103.26,'JPY_2021': 114.41,'JPY_2022': 132.76,'JPY_2023': 139.46} #Yenes el dolar
cambiosAnhoPYG = {'PYG_2018': 5983.90,'PYG_2019': 6368.68,'PYG_2020': 6750.56,'PYG_2021': 6881.53,'PYG_2022': 7364.29,'PYG_2023': 7252.55} #guaranies el dolar
cambiosAnhoEUR = {'EUR_2018': 0.87,'EUR_2019': 0.89,'EUR_2020': 0.81,'EUR_2021': 0.88,'EUR_2022': 0.93,'EUR_2023': 0.92}
cambiosAnhoCAD = {'CAD_2018': 1.36,'CAD_2019': 1.31,'CAD_2020': 1.29,'CAD_2021': 1.26,'CAD_2022': 1.36,'CAD_2023': 1.33}
cambiosAnhoZAR = {'ZAR_2018': 14.43,'ZAR_2019': 14.02,'ZAR_2020': 14.40,'ZAR_2021': 15.95,'ZAR_2022': 17.02,'ZAR_2023': 18.26}
```

```python
# Funcion para cambiar todas las monedas a Dolares Abdo
def convertirMonedaAbdo(aGov, cambio):
    for key, value in cambio.items():
        currency, year = key.split('_')
        aGov.loc[(aGov['monedaBasePrestamo'] == currency) & (aGov['anioCorte'] == int(year)), 'montoPrestamo'] /= value
        aGov.loc[(aGov['monedaBasePrestamo'] == currency) & (aGov['anioCorte'] == int(year)), 'monedaBasePrestamo'] = 'USD'
    return aGov
```

```python
# guardo un diccionario para los diccionarios de cambios
cambios = [cambiosAnhoJPY, cambiosAnhoPYG, cambiosAnhoEUR, cambiosAnhoCAD, cambiosAnhoZAR]

# ciclo para cambiar entre todas las monedas
for i in range(0, 5):
    cambio = cambios[i]
    df_convertido = convertirMonedaAbdo(aGov, cambio)
```
